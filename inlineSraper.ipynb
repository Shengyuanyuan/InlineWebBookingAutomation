```python
# 掛載瀏覽器
driver = webdriver.Chrome('./chromedriver')#options=op

def bookingAutomation():
    # 進入指向的url
    driver.get('https://inline.app/booking/-L31G0d9splHSo6q4jEp/-L31G0d9splHSo6q4jEq')
    # driver.get('https://inline.app/booking/-MkBSdjj_81Mjn1vAZA6:inline-live-2/-MkBSdrkF0niJoOgE4yz?fbclid=IwAR1-XCiOBbDzF_6Umxvd-78CHP4UNI2uf4EYRNY5UEy_VVEYOqSZCeSfnrE')

    # canStop為Boolean，用來停止無窮迴圈
    canStop = False
    while canStop == False:
        try:
            # 監聽可選的時間
            element = WebDriverWait(driver, 2).until(
                EC.presence_of_element_located((By.XPATH, "/html/body/div[1]/div/div/main/div[3]/div[2]/div[2]/button[2]"))
            )
            canStop = True
        
        # 拋出timeOut例外，繼續無窮迴圈，並刷新網站(針對Client side render)
        except TimeoutException:
            driver.refresh()

    # 移動至可選擇的按鈕畫面
    driver.execute_script('window.scrollTo(0, 800);')
    sleep(0.3)

    # 選擇用餐人數
    driver.find_element_by_xpath('/html/body/div[1]/div/div/main/div[3]/div[1]/div[1]/div/select[1]/option[4]').click() # 3人
    # sleep(0.2)

    # Line(1)為點開日期的select bar, Line(2)選擇日期, Line(3)選擇時間
    driver.find_element_by_xpath('/html/body/div[1]/div/div/main/div[3]/div[1]/div[2]/div').click()
    sleep(0.6)
    driver.find_element_by_xpath('/html/body/div[1]/div/div/main/div[3]/div[1]/div[3]/div[1]/div[3]/div[2]/div[6]/div[2]/span').click()
    sleep(0.5)
    driver.execute_script('window.scrollTo(0, 1500);')
    sleep(0.5)
    driver.find_element_by_xpath('/html/body/div[1]/div/div/main/div[3]/div[2]/div[2]/button[2]').click()
    sleep(0.1)

        
    # 下一步，填寫資料
    driver.find_element_by_xpath('/html/body/div[1]/div/div/main/div[4]/div[2]/button').click()

bookingAutomation()

# 監聽填寫資料頁面(第二個頁面)
try:
    element = WebDriverWait(driver, 1000).until(
        EC.presence_of_element_located((By.ID, "contact-form"))
    )

# 拋出例外，回到最一開始的頁面，重新搶票
except TimeoutException:
    bookingAutomation()


# 載入貼寫資料後，開始自動填寫資料
finally:
    # 姓名
    user = driver.find_element_by_id('name')
    sleep(0.2)
    user.send_keys(" ") # 輸入你的名字
    sleep(0.2)
    driver.find_element_by_id('gender-male').click()
    sleep(0.2)

    # 手機
    phone = driver.find_element_by_id('phone')
    phone.send_keys(" ") # 輸入你的手機號碼
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    sleep(0.5)

    # 勾選checkbox
    sleep(0.5)
    driver.find_element_by_css_selector("#contact-form > section > div.sc-bdvvtL.itsbag > div > div:nth-child(5)").click() #朋友聚會選項
    driver.find_element_by_css_selector("#contact-form > div.sc-bdvvtL.icJJQV > div.sc-bdvvtL.kjGxzc > label").click() #同意事項說明
    sleep(0.25)


    #送出訂單
    # driver.find_element_by_xpath('/html/body/div[1]/div/div/form/div[2]/button[1]').click()
    canStop = True



```
